```python
# 🧼 Day 4 – Pandas Mini Project: Airbnb Athens Cleaning

import pandas as pd

# 📄 Διαβάζουμε το αρχείο
df = pd.read_csv("../data/raw/airbnb_athens.csv")
print("🔍 Αρχικές γραμμές:\n", df)

# ℹ️ Πληροφορίες
print("\n📋 Πληροφορίες:")
print(df.info())

# 🧠 Περιγραφικά στατιστικά
print("\n📊 Στατιστικά:")
print(df.describe())

# 🔎 Έλεγχος για ελλιπείς τιμές (nulls)
print("\n❓ Null values per column:")
print(df.isnull().sum())

# 🗑️ Διαγραφή γραμμών με null τιμές
df_cleaned = df.dropna()

# Εναλλακτικά: γέμισμα nulls (σε τιμές)
# df_filled = df.fillna({"price": df["price"].mean(), "beds": 1})

# 🧮 Ομαδοποίηση κατά περιοχή
grouped = df_cleaned.groupby("neighborhood").agg({
    "price": "mean",
    "beds": "mean"
}).reset_index()

print("\n📊 Μέσες τιμές ανά περιοχή:\n", grouped)

# 💾 Αποθήκευση καθαρισμένου dataset
df_cleaned.to_csv("../data/processed/airbnb_athens_cleaned.csv", index=False)
print("\n✅ Αποθηκεύτηκε το cleaned αρχείο.")
```
