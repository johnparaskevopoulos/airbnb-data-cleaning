```python
# 📊 Day 5 – Plotting με Pandas & Matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# 🔄 Φόρτωση καθαρισμένων δεδομένων
df = pd.read_csv("../data/processed/airbnb_athens_cleaned.csv")

# 🔁 Ομαδοποίηση ανά περιοχή
grouped = df.groupby("neighborhood").agg({
    "price": "mean",
    "beds": "mean"
}).reset_index()

# 📊 Bar Chart – Μέση Τιμή Airbnb ανά Περιοχή
plt.figure(figsize=(8, 4))
plt.bar(grouped["neighborhood"], grouped["price"], color='skyblue')
plt.title("Μέση Τιμή Airbnb ανά Περιοχή")
plt.xlabel("Περιοχή")
plt.ylabel("Τιμή (€)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../reports/price_per_area.png")
plt.show()

# 🛏️ Bar Chart – Μέσος Όρος Κρεβατιών ανά Περιοχή
plt.figure(figsize=(8, 4))
plt.bar(grouped["neighborhood"], grouped["beds"], color='lightgreen')
plt.title("Μέσος Αριθμός Κρεβατιών ανά Περιοχή")
plt.xlabel("Περιοχή")
plt.ylabel("Κρεβάτια")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../reports/beds_per_area.png")
plt.show()

# 🧁 Pie Chart – Ποσοστά Καταχωρήσεων Airbnb ανά Περιοχή
plt.figure(figsize=(6, 6))
area_counts = df["neighborhood"].value_counts()
plt.pie(area_counts, labels=area_counts.index, autopct='%1.1f%%')
plt.title("Ποσοστά Καταχωρήσεων Airbnb")
plt.savefig("../reports/airbnb_pie_chart.png")
plt.show()

```
