```python
# 🐼 Day 3 – Pandas Basics: Καθαρισμός & Ανάλυση Δεδομένων

import pandas as pd

# 📄 Διαβάζουμε ένα dataset (π.χ. test.csv από χθες)
df = pd.read_csv("../data/test.csv")

# 🔍 Γενική εικόνα του dataset
print(df.head())          # Πρώτες γραμμές
print(df.info())          # Τύποι δεδομένων
print(df.describe())      # Στατιστικά περιγραφικά

# 🧹 Φιλτράρισμα γραμμών
filtered = df[df["value"] > 10]
print("Τιμές > 10:\n", filtered)

# 📊 Ταξινόμηση
sorted_df = df.sort_values(by="value", ascending=False)
print("Κατά φθίνουσα σειρά:\n", sorted_df)

# 🔁 groupby (θα χρειαστεί μεγαλύτερο dataset για καλύτερο παράδειγμα)
# grouped = df.groupby("device_id")["value"].mean()

# 🩺 Αντιμετώπιση ελλιπών τιμών
df_with_nan = df.copy()
df_with_nan.loc[1, "value"] = None  # Προσθέτουμε μια NaN τεχνητά

print("\nΑρχείο με NaN:")
print(df_with_nan)

# Διαγραφή NaN
print("\nΧωρίς NaN:")
print(df_with_nan.dropna())

# Αντικατάσταση NaN με 0
print("\nΑντικατάσταση NaN με 0:")
print(df_with_nan.fillna(0))
```
